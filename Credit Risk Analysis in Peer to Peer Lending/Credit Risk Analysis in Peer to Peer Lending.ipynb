# Credit Risk Analysis in Peer to Peer Lending

# Business Understanding: 

Online Peer-to-Peer Lending has been on raise with both borrowers and lenders reaping profits over approaching banks for saving and borrowing, respectively. I want to do some research on this Micro â€“ financing feature - Peer to Peer Lending in banking domain. Peer to Peer lending is considered as a risky but high return business. Lenders get better returns by lending to borrowers at a higher interest rate than what Banks offer for saving their money and for borrowers it gives them flexibility and cheaper rate than banks who check credit scores from credit reporting agencies before offering loans. 


## Business Goal: 

Traditional lending offered by banks generally verify some sort of FICO score from credit reporting agency before finalizing whether to offer loan to the customer or not and if ok to offer, what would be the loan interest rate etc. But in Peer to Peer lending borrowers cannot base their decisions on credit reporting agency, as most of their customers turn out to be high risk any way. So, P2P lending needs different strategy to identify potential good borrowers and distinguish them defaulting borrowers. 



#### Objective 1:

	Customer segmentation into good borrower or potential defaulter, there by avoiding the risk of lending him/her.

#### Objective 2:

 	Identifying optimum interest rate that can be charged to the customers based on their profile.

```python
def month_to_int(Month):
    #print(Month)
    mmm = Month.lower()
    if mmm == 'dec':
        return 12
    if mmm == 'nov':
        return 11
    if mmm == 'oct':
        return 10
    if mmm == 'sep':
        return 9
    if mmm == 'aug':
        return 8
    if mmm == 'jul':
        return 7
    if mmm == 'jun':
        return 6
    if mmm == 'may':
        return 5
    if mmm == 'apr':
        return 4
    if mmm == 'mar':
        return 3
    if mmm == 'feb':
        return 2
    if mmm == 'jan':
        return 1
    
def str_to_date(datestr):

    yr = datestr[4:8]
    mn = month_to_int(datestr[0:3])

    crtd_date = str(yr) +'-'+ str(mn) +'-' + str(1)
    return (dt.strptime(crtd_date, "%Y-%m-%d"))
```

```python
def conv_subgrade(text):
    
    if (str(text) == 'nan') : #text.isna() : #isnull() : #len(text) == 0:
        return 'N'
    else :
        firstchr =  (''.join(filter(str.isalpha, text)))
        #print(firstchr)
        if firstchr == 'A':
            first = 7
        if firstchr == 'B':
            first = 6
        if firstchr == 'C':
            first = 5
        if firstchr == 'D':
            first = 4
        if firstchr == 'E':
            first = 3
        if firstchr == 'F':
            first = 2
        if firstchr == 'G':
            first = 1

        
        second = (''.join(filter(str.isdigit, text)))
        exp = str(first) + str(second)
        return (exp)
```

#### Feature Selection by finding variable importance

As our objective is to identify the risky loans, in other words we need to classify whether a particular loan could be defaulted or fully paid. Target variable here is "loan_status_cd". Rest all could be our features.

There are too many variables to make this prediction, but not all will provide us insights into this prediction. So, I will use ExtraTreesClassifier to select the important features here. 

```python
features = accept_df_2014_2015_v6.loc[ : , accept_df_2014_2015_v6.columns != 'loan_status_cd' ]

features.head()
```

```python
target =  accept_df_2014_2015_v6['loan_status_cd']
```

```python
# feature selection
fsmodel = ExtraTreesClassifier(n_estimators=10)
fsmodel.fit(features, target)
impfeatures = fsmodel.feature_importances_
print(impfeatures)
```

```python
print(features.columns)
```

```python
def fn_multiclass_metrics(actual_label, predicted_label):
    """
    function that takes acutal labels and predicted labels and returns
    accuracy, auc, precision, recall and f1 scores
    average = 'weighted' for multi class classification
    """
    accuracy = accuracy_score(actual_label, predicted_label)
    precision = precision_score(actual_label, predicted_label, average = 'weighted')
    recall = recall_score(actual_label, predicted_label, average = 'weighted')
    f1 = f1_score(actual_label, predicted_label, average = 'weighted')

    return (accuracy, precision, recall, f1)
```

```python
def plotroc(false_positive_rate, true_positive_rate):

    """
    A function that takes false positive rate and true positive rate for plotting ROC curve
    Returns: none
    Expected o/p: ROC plot
    """

    # Plot ROC curve
    plt.figure(figsize=(12, 8))
    plt.title("Receiver Operating Characteristic", fontdict=titlefont)
    plt.plot(false_positive_rate, true_positive_rate,  label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], ls="--")
    plt.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
    plt.ylabel("True Positive Rate" , fontdict=labelfont)
    plt.xlabel("False Positive Rate", fontdict=labelfont)
    plt.legend(loc="lower right")
    plt.show()
```
